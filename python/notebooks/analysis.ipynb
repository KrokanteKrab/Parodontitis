# Data from csv

```python
dfs = [
    pd.read_csv("../data/generated/patients-v5.csv"),
    pd.read_csv("../data/synthetic/patients-B-v1.csv"),
    pd.read_csv("../data/synthetic/patients-C-v1.csv")
]
```

```python
for df in dfs:
    df = df[[
        # 'PATIENT_ID',
        'GENDER_MALE',
        'GENDER_FEMALE',
        # 'BIRTH_DATE',
        'AGE_RANGE_20',
        'AGE_RANGE_40',
        'AGE_RANGE_60',
        # 'VISIT_DATE',
        'TREATING_PROVIDER_DENTIST',
        'TREATING_PROVIDER_FACULTY',
        'TREATING_PROVIDER_STUDENT',
        'PROCEDURE_A',
        'PROCEDURE_B',
        'BLEEDING_ON_PROBING',
        'NR_OF_POCKET',
        'NR_OF_FURCATION',
        'NR_OF_MOBILITY',
        'TOTAL_LOSS_OF_ATTACHMENT_LEVEL',
        'HAS_PARODONTITIS',
        # 'NOICE_MODIFIED',
        # 'STUDENT_ERROR',
        # 'TRUE_HAS_PARODONTITIS'
    ]]

    # Visualizing a Pandas Correlation Matrix Using Seaborn
    matrix = df.corr().round(2)
    sns.set(rc={'figure.figsize':(15,6)})
    sns.heatmap(matrix, annot=True)
    plt.show()
```
